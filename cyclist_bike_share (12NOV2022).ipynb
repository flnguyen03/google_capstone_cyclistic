With the Geo coordinate data, I implemented the [Haversine function](insert link). This helps me practice on creating and storing data in the arrays versus importing another python library.

```python
import math
# Creating a Calculation for Distance Between GPS Points in Python
def haversine(coord1, coord2):
    R = 6371000  # Earth radius in meters
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    
    phi1, phi2 = math.radians(lat1), math.radians(lat2) 
    dphi       = math.radians(lat2 - lat1)
    dlambda    = math.radians(lon2 - lon1)
    
    a = math.sin(dphi/2)**2 + \
        math.cos(phi1)*math.cos(phi2)*math.sin(dlambda/2)**2
    
    return 2*R*math.atan2(math.sqrt(a), math.sqrt(1 - a))
```

```python
arr1 = np.array(df_tna['start_lat']).reshape([-1,1])
arr2 = np.array(df_tna['start_lng']).reshape([-1,1])
start_location = list(map(lambda coord: [coord[0][0], coord[1][0]], zip(arr1,arr2)))

arr3 = np.array(df_tna['end_lat']).reshape([-1,1])
arr4 = np.array(df_tna['end_lng']).reshape([-1,1])
end_location = list(map(lambda coord: [coord[0][0], coord[1][0]], zip(arr3,arr4)))
```

```python
list = []

for i,j in zip(start_location,end_location):
    distance = haversine(i,j)
    list.append(distance)

df_d = pd.DataFrame(list, columns= ['ride_distance'])
print("Created DataFrame:\n", df_d, "\n")
```
